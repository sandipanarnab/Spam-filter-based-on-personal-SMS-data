```python
data_filtered_en= data_filtered[data_filtered['Language']== 'en']

data_filtered_en.head()
```

<span style="color: orange;"> Step 1: Data-preprocessing</span>

<span style="color: yellow;">First let us extract all the Statitical features such as length of the messages, number of alphanumeric caharcters, number of spaces, number of numerical digits from each messages as these could be sueful features for building a cross-langauge spam filter.</span>

```python
def extract_features(df):

    df = df.copy()
    
    df.loc[:, 'message_length'] = df.loc[:, 'Received'].apply(len)
    
    df.loc[:, 'num_digits'] = df.loc[:, 'Received'].apply(lambda x: sum(c.isdigit() for c in x))
    
    df.loc[:, 'num_special_chars'] = df.loc[:, 'Received'].apply(lambda x: sum(not c.isalnum() for c in x))

    df.loc[:, 'num_spaces'] = df.loc[:, 'Received'].apply(lambda x: sum(c.isspace() for c in x))
    
    return df

data_filtered_en = extract_features(data_filtered_en)

data_filtered_en.head()
```

<span style="color: yellow;">Second let us remove the stop-words, extra spaces and convert all texts into lower case.</span>

```python
# Text cleaning function
def clean_text(text):
    text = re.sub(r'\W', ' ', text)  # Remove non-alphanumeric characters (including punctuation)
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.lower()  # Convert to lowercase
    return text

data_filtered_en.loc[:, 'cleaned_message'] = data_filtered_en.loc[:, 'Received'].apply(clean_text)

data_filtered_en.head()
```

<span style="color: orange;">Step 2. Feature Selection, lemmatization</span>

<span style="color: yellow;">First we will try to build a base model based on NLP. For that we will discard all the unnecessary columns.</span>

```python
data_filtered_en= data_filtered_en.drop(columns= ['Phone Number', 'Received', 'Message', 'Language', 'message_length', 'num_digits', 'num_special_chars', 'num_spaces'], axis= 1)

data_filtered_en.head()
```

```python
data_filtered_en.info()
```

```python
nlp = spacy.load('en_core_web_sm')

def preprocess_text(text):
    doc = nlp(text)
    return ' '.join([token.lemma_ for token in doc])

# Apply preprocessing
data_filtered_en['processed_message'] = data_filtered_en['cleaned_message'].apply(preprocess_text)

data_filtered_en= data_filtered_en.reset_index(drop=True)

data_filtered_en.head()
```

```python
data_filtered_en['Spam']= data_filtered_en['Spam'].astype('int64')

data_filtered_en.info()

data_filtered_en.head()
```

```python

data_filtered_en= data_filtered_en.drop(columns= ['cleaned_message'], axis=1)

data_filtered_en.head()
```

```python
import tensorflow as tf
import gc
from transformers import DistilBertTokenizer, TFDistilBertForSequenceClassification
from sklearn.metrics import classification_report

# Enable mixed precision
policy = tf.keras.mixed_precision.Policy('mixed_float16')
tf.keras.mixed_precision.set_global_policy(policy)

# Clear TensorFlow session
tf.keras.backend.clear_session()

# Collect garbage
gc.collect()

# Define and Train the DistilBERT Model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
MAX_LEN = 96  # You can adjust this based on your text length

def encode_texts(texts):
    return tokenizer(
        texts.tolist(),
        max_length=MAX_LEN,
        padding='max_length',
        truncation=True,
        return_tensors='tf'
    )

encoded_texts = encode_texts(data_filtered_en['processed_message'])

# Create TensorFlow dataset
dataset = tf.data.Dataset.from_tensor_slices((
    dict(encoded_texts),
    tf.cast(data_filtered_en['Spam'].values, tf.int32)
))

# Split dataset
train_size = int(0.8 * len(data_filtered_en))
train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)

BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

def prepare_dataset(features, labels):
    return {
        'input_ids': features['input_ids'],
        'attention_mask': features['attention_mask']
    }, labels

train_dataset = (train_dataset
                 .map(prepare_dataset, num_parallel_calls=AUTOTUNE)
                 .shuffle(len(data_filtered_en))
                 .batch(BATCH_SIZE)
                 .prefetch(AUTOTUNE))

val_dataset = (val_dataset
               .map(prepare_dataset, num_parallel_calls=AUTOTUNE)
               .batch(BATCH_SIZE)
               .prefetch(AUTOTUNE))

# Define and Train the DistilBERT Model
model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-6)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

# Early stopping callback
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=100,  # Increase epochs, early stopping will prevent overfitting
    callbacks=[early_stopping]
)

# Evaluate the Model
loss, accuracy = model.evaluate(val_dataset)
print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")

# Generate predictions
predictions = model.predict(val_dataset)
pred_labels = tf.argmax(predictions.logits, axis=1).numpy()

# Get true labels
val_labels = tf.concat([labels for _, labels in val_dataset], axis=0).numpy()

# Classification report
print(classification_report(val_labels, pred_labels))
```

```python
import os

def save_model_and_tokenizer(model, tokenizer, save_directory, model_name):
    """
    Save the model and tokenizer to the specified directory with the given model name.
    
    Args:
        model: The trained DistilBERT model to be saved.
        tokenizer: The tokenizer associated with the model.
        save_directory (str): The base directory where the model and tokenizer will be saved.
        model_name (str): The name of the model.
    """
    # Define the full save path
    full_save_path = os.path.join(save_directory, model_name)
    
    # Ensure the save directory exists
    if not os.path.exists(full_save_path):
        os.makedirs(full_save_path)
    
    # Save the model and tokenizer
    model.save_pretrained(full_save_path)
    tokenizer.save_pretrained(full_save_path)
    
    print(f"Model and tokenizer saved to {full_save_path}")

# Example usage
save_directory = "/mnt/c/Users/sandi/Desktop/My Working Git/Spam filter based on personal SMS data/models"
model_name = "distilBERT_gpu_64_16"
save_model_and_tokenizer(model, tokenizer, save_directory, model_name)
```
