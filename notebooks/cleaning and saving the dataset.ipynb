```python
columns_to_delete = ['ID', 'Name', 'Date Time', 'Sent', 'Status', 'Unnamed: 8']

# Delete the specified columns
data_filtered = data_filtered.drop(columns=columns_to_delete, axis=1)

data_filtered.head(20)
```


<span style="color: yellow;"> As we can see there are a few Non-English caharacters in the Received column and we need to either filter out the English characters from each rows of that column or we can delete those rows.</span>

```python
import pandas as pd
import unicodedata

# Assuming you have the DataFrame 'df'

def remove_non_english(text):
    return ''.join(char for char in text if ord(char) < 128)

# Remove non-English characters from the 'column_name' column
data_filtered['Received'] = data_filtered['Received'].apply(remove_non_english)

data_filtered.head(20)
```

```python
data_filtered.reset_index(drop=True, inplace=True)
data_filtered.head(20)
```

```python
data_filtered.shape
```


<span style="color: yellow;"> As the dataframe rows may contain sensitive information in numbers which may be related to privacy and security we will replace the numbers with 0000.</span>

```python

import re

def replace_numbers_with_0000(text):

    # The regular expression pattern to match all numeric sequences
    
    pattern = r'\d+'
    
    # Replace the matched pattern with '0000'
    return re.sub(pattern, '0000', text)

# Apply the function to the 'Received' column
data_filtered['Received'] = data_filtered['Received'].apply(replace_numbers_with_0000)

data_filtered.head(20)
```

```python
data_filtered.drop_duplicates(subset='Received', inplace=True)

data_filtered.head(20)
```
